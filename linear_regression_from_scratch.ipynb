```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([40, 50, 60])
y = np.array([80,100, 120])

w = 0
b = 0
learning_rate = 0.0001
epochs = 1000
loss_history = []

for epoch in range(epochs):
    y_pred = w*x + b
    loss = np.mean((y_pred-y)**2)

    dw = np.mean(2 * (y_pred-y)*x)
    db = np.mean(2 * (y_pred-y))

    w -= learning_rate * dw
    b -= learning_rate * db
```

```python
print(f'\nfinal: w = {w:.4f}, b = {b:.4f}')
```

```python
x_new = 55
y_pred = w * x_new + b
print(f'pred valie {x_new} м²: {y_pred:.2f} k. $')
print()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(x, y, color='blue', s=100, label='Training data')
plt.plot(x, w * x + b, color='red', linewidth=2, label='Regression line')
plt.scatter([x_new], [y_pred], color='green', s=200, marker='*',
           label=f'Prediction ({x_new} m²)')
plt.xlabel('Area (m²)', fontsize=12)
plt.ylabel('Price (thousand $)', fontsize=12)
plt.title('Linear Regression: Price vs Area', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
```
